```python
#import the required libraries
import pandas as pd 
import seaborn as sns 
import numpy as np
```

```python
#import tips dataset from seaborn 
tip = sns.load_dataset("tips")
```

```python
tip.head()
```

```python
#get the shape of the data
tip.shape
```

```python
#view the top 20 row 
tip.head(20)
```

```python
#view the bottom 20 row 
tip.tail(20)
```

```python
#Get the avg total_bill by only male
grouped = tip.groupby(by = 'sex').agg({'total_bill':'mean'})
grouped.loc['Male']
```

```python
#Get the avg(mean,median) total_bill and tip given by smoker and non smoker
tip.groupby(by='smoker').agg({'total_bill':['mean','median'], 'tip':['mean','median']}).reset_index()
```

```python
# what is most frequent time of vist
tip['time'].value_counts().head(1)
```

```python
#Which day customers are giving higest tips
tip.groupby(by='day').agg({'tip':'sum'}).sort_values('tip', ascending = False).head(1)
```

```python
#get the avg(median) tip and total_bill amount by smoker(yes) on sunday
tip[tip['day'] == 'Sun'].agg({'tip':'median','total_bill':'median'})
```

```python
#1 get the (mean,median,min,max) tip and total_bill amount by smoker(yes) on friday with time(Lunch)

df = tip[(tip['smoker'] == 'Yes') & (tip['time'] == 'Lunch') & (tip['day'] == 'Fri')]
df.agg({'total_bill':['mean','median','min','max'],'tip':['mean','median','min','max']})
```

```python
#2 What is min bill by a smoker on sunday

df = tip[(tip['smoker'] == 'Yes') & (tip['day']=='Sun')]
df['total_bill'].min()
```

```python
#3 what is max tip by a non smoker on thursday

df = tip[(tip['smoker'] == 'No') & (tip['day']=='Thur')]
df['tip'].max()
```

```python
#4 what is total_bill made on sat by non smokers

df = tip[(tip['smoker'] == 'No') & (tip['day']=='Sat')]
df['total_bill'].sum()
```

```python
#which gender is having highest bill amount
tip.groupby('sex')['total_bill'].max()
```

```python
#Get the min, max, and avg(mean, and median) of total bill and tip
tip.agg({'total_bill':['mean','median'],'tip':['mean','median']})
```

```python
#create a new column total_bill amount category(low, moderate, high)
tip['bill_category'] = pd.cut(tip['total_bill'], bins = 3, labels = ['low','moderate','high'])
tip.head()
```
